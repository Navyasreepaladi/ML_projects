```python
# importing required modules
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# reading the daimond dataset 
path = '/content/diamonds.csv'
df = pd.read_csv(path,index_col=0)

df.head()
```

```python
# converting categorical variables into numerical
object_cols = ['cut', 'color', 'clarity']
label_encoder = LabelEncoder()
for col in object_cols:
    df[col] = label_encoder.fit_transform(df[col])
df.head()
```

```python
X = df.drop("price", axis = "columns")
y = df["price"]

# splitting data into train (70%), validation (20%), test (10%) 
X_train, X_, y_train, y_ = train_test_split(X, y, test_size=0.3)
X_test, X_val, y_test, y_val = train_test_split(X_, y_, test_size=0.6)

print('train shape:', X_train.shape, y_train.shape)
print('validation shape:', X_val.shape, y_val.shape)
print('test shape:', X_test.shape, y_test.shape)
```

```python
# training and Prediction using Linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_val)
print("MSE :", mean_squared_error(y_val,y_pred))
print("MAE :", mean_absolute_error(y_val,y_pred))
```

```python
# training and Prediction using Decision regression model
model1 = DecisionTreeRegressor()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_val)
print("MSE :", mean_squared_error(y_val,y_pred1))
print("MAE :", mean_absolute_error(y_val,y_pred1))
```

```python
# Predicting Price for test data
preds = model.predict(X_test)
preds1 = model1.predict(X_test)
print('prediction using linear regression\n',preds)
print('\nprediction using decision tree regression\n',preds1)
```
